```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.activations import sigmoid
```

```python
Size = np.array([1200, 3000])
Price = np.array([100, 250])


fig, ax = plt.subplots(1,1)
ax.scatter(Size, Price, marker='x', c='r', label="Data Points")
ax.legend( fontsize='xx-large')
ax.set_ylabel('Price (in 1000s of dollars)', fontsize='xx-large')
ax.set_xlabel('Size (1000 sqft)', fontsize='xx-large')
plt.show()
```

```python
linear_layer = tf.keras.layers.Dense(units=1, activation = 'linear', )
linear_layer.get_weights() ## no weights intially
linear_layer(Size[0].reshape(1,1))
linear_layer.get_weights() ## random weights 

set_w = np.array([[200]])
set_b = np.array([100])

# set_weights takes a list of numpy arrays
linear_layer.set_weights([set_w, set_b])
print(linear_layer.get_weights())

result = linear_layer(Size[0].reshape(1,1))
print(result)
np_result = np.dot(Size[0].reshape(1,1), set_w) + set_b
print(np_result)
```

```python
## lets pack a single layer in to model and call predict function
model = tf.keras.Sequential([linear_layer])

# Pass input data through the model using the predict method
result = model.predict(Size[0].reshape(1, 1))
print(result)
```
